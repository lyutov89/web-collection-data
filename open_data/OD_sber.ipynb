```python
name = df['name'] == df.name.unique()[i]
region = df['region'] == df.region.unique()[j]
date = (df['date'] > df.date.unique()[4]) & (df['date'] < df.date.unique()[72])

graph = df[name & region & date]
#print(result)


plt.subplots(figsize=(16, 8))
plt.style.use('fivethirtyeight')
labeldict={'fontsize': '14', 'color': "#000000", 'family': "Arial" }
plt.plot(graph.date, graph.value, color="green")
plt.ylabel('value')
plt.xlabel('time')
plt.title(
    f'{df.name.unique()[i]} в {df.region.unique()[j]}',
    fontsize=16,
    fontweight="bold",
    color="#000000",
    family="Arial",
)

plt.show()
```
